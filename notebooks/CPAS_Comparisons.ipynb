# CPAS Comparisions
This notebook compares Grant Table Schemas across applications years.

This notebook depends on Oracle SQL Developer Exports of schema metadata.
The files are now stored in `data/grnt_exports`.

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import xlrd, xlwt, xlsxwriter
```

```python
root_path = "../data/grnt_exports/g"
```

```python
dfs = []
for i in range(37,46):
    df = pd.read_excel(root_path+str(i)+".xlsx", index_col = None)
    dfs.append(df)
```

```python
tbl_container = []
for df in dfs:
    tbl_names = df.TABLE_NAME.unique()
    tbl_container.append(tbl_names)
common_tbls = set.intersection(*map(set,tbl_container))
print(common_tbls)
```

```python
print(f'number of common tables during last 10 Grants cycles is: {len(common_tbls)}')
```

```python
lookups = 'GRNT_L_'
for i in filter(lambda i: i.startswith(lookups), common_tbls):
          print(i)
```

```python
for j in common_tbls:
    if not j.startswith('GRNT_L_'):
        print(j)
```
