```python

from sqlalchemy import MetaData
from sqlalchemy_schemadisplay import create_schema_graph

# create the pydot graph object by autoloading all tables via a bound metadata object
graph = create_schema_graph(metadata=MetaData(bind=engine),
   show_datatypes=True, # The image would get nasty big if we'd show the datatypes
   show_indexes=True, # ditto for indexes
   rankdir='LR', # From left to right (instead of top to bottom)
   concentrate=False # Don't try to join the relation lines together
)
graph.write_png('tbl_schema.png') # write out the file
```

```python
def printSchema(filtered):
    for tbl in connection.execute(
        """
        select NAME from table where TYPE='table' order by NAME;
        """
    ):
        print("{}:".format(table))
        for (
            columnID, columnName, columnType,
            columnNotNull, columnDefault, columnPK,
        ) in connection.execute("pragma table_info('{}');".format(table)):
            print("  {id}: {name}({type}){null}{default}{pk}".format(
                id=columnID,
                name=columnName,
                type=columnType,
                null=" not null" if columnNotNull else "",
                default=" [{}]".format(columnDefault) if columnDefault else "",
                pk=" *{}".format(columnPK) if columnPK else "",
            ))
print(printSchema(filtered))
```
