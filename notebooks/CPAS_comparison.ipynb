# CPAS Comparision
This notebook compares Grant Table Schemas across applications years.

Notebook depend on Oracle SQL Developer Exports
Those are now stored in `data/grnt_exports`

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import xlrd, xlwt, xlsxwriter
```

```python
root_path = "../data/grnt_exports/g"
```

```python
dfs = []
for i in range(30,46):
    df = pd.read_excel(root_path+str(i)+".xlsx", index_col = None)
    dfs.append(df)
```

```python

tbl_container = []
for df in dfs:
    tbl_names = df.TABLE_NAME.unique()
    tbl_container.append(tbl_names)
common_tbls = set.intersection(*map(set,tbl_container))
print(common_tbls)
```

```python
print(f'number of common tables is: {len(common_tbls)}')
```
